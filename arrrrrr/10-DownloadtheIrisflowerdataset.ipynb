```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load the Iris dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, names=column_names)

# Show the first few rows
print(df.head())
```

```python

# 1.List the features and their types
feature_types = df.dtypes
print("Features and their types:")
print(feature_types)
```

```python
# 1. Display dataset info
print("Dataset Info:")
print(df.info())
```

```python
# Data Preprocessing
# 1. Check for Missing Values
# We can start by checking for any missing values in the dataset:


# Check for missing values
print(df.isnull().sum())
```

```python

# How to Read a Boxplot:
#    Min     Q1       Median     Q3     Max
#           [   IQR = Q3 - Q1   ]

# The box shows the middle 50% of the data (from Q1 to Q3).
# Minimum (excluding outliers)

# First Quartile (Q1) – 25% of the data is below this point

# Median (Q2) – Middle value of the dataset

# Third Quartile (Q3) – 75% of the data is below this point

# Maximum (excluding outliers)

# Outliers – Data points that fall outside the usual range (shown as dots or stars)


```

```python
 # Normalize the Data
# Normalization is important when using algorithms that are sensitive to the scale of data (like KNN, SVM, etc.). We'll use MinMax Scaling to scale the data:


from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Normalize the numeric features (excluding the 'species' column)
df_scaled = pd.DataFrame(scaler.fit_transform(df.drop(columns='species')), columns=df.columns[:-1])

# Show the normalized data
print(df_scaled.head())
```

```python
# Feature Distribution – Histograms and Boxplots
# 2. Histograms
# Histograms are great to visualize the distribution of each feature.


# Create a histogram for each feature
df_scaled.hist(figsize=(12, 8))
#  this will scale from range 0-1.0
plt.tight_layout()
plt.show()
```

```python
# Boxplot
# A boxplot will give you a good indication of the distribution and potential outliers for each feature:


# 3.Boxplot for each feature
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.drop(columns='species'))
plt.title('Boxplot for Feature Distribution')
plt.show()
```
