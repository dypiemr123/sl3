```python
# Problem Statement: 10 Data Visualization III
# Download the Iris flower dataset or any other dataset into a DataFrame.
# (e.g., https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and
# give the inference as:
# 1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
# 2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
# 3. Create a boxplot for each feature in the dataset.
# 4. Compare distributions and identify outliers.

# Importing required libraries
import pandas as pd               # For data handling
import seaborn as sns             # For visualization (boxplots)
import matplotlib.pyplot as plt   # For plotting histograms and other charts
from sklearn.preprocessing import MinMaxScaler  # For normalization

# Step 1: Load the Iris dataset from the UCI repository
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# Define the column names as the original dataset doesn't have headers
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Read the dataset into a pandas DataFrame
df = pd.read_csv(url, names=column_names)

# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Step 2: List the features and their data types
print("\nFeature Names and Their Data Types:")
print(df.dtypes)

# Display dataset info including data types and non-null counts
print("\nDataset Info:")
print(df.info())

# Step 3: Check for missing values in each column
print("\nMissing values in each column:")
print(df.isnull().sum())  # No missing values expected in the Iris dataset

# Step 4: Normalize the numeric features using Min-Max Scaling (0 to 1 scale)
scaler = MinMaxScaler()  # Initialize the scaler

# Apply scaler to numeric columns only (exclude 'species')
df_scaled = pd.DataFrame(scaler.fit_transform(df.drop(columns='species')), columns=column_names[:-1])

# Display the normalized dataset
print("\nFirst 5 rows after normalization (0 to 1 scale):")
print(df_scaled.head())

# Step 5: Create Histograms for all numeric features
print("\nPlotting Histograms for each feature...")
df_scaled.hist(figsize=(12, 8), bins=15)  # Adjust number of bins for clarity
plt.suptitle('Histogram of Scaled Features')
plt.tight_layout()
plt.show()

# Step 6: Create Boxplots to visualize feature distribution and detect outliers
print("Plotting Boxplots for feature distribution...")
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.drop(columns='species'))  # Use original data to observe real values
plt.title('Boxplot for Each Feature (Detecting Outliers)')
plt.ylabel('Measurement in cm')
plt.show()

# Step 7: Observations (to be written separately or printed below if required)
# - Histograms help understand the frequency distribution of each feature.
# - Boxplots reveal:
#     * Potential outliers (points outside whiskers)
#     * Skewness and median positions
#     * Variability (IQR width)
# - Petal features have greater variation than sepal features.
# - Sepal width has some potential outliers on both ends.

# Note: These observations should ideally be documented in a report or markdown cell if in Jupyter.
```
