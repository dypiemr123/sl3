```python
# 🔍 Problem Statement:
# 5. Data Analytics II
# 1. Implement logistic regression using Python/R to perform classification on Social_Network_Ads.csv dataset.
# 2. Compute Confusion matrix to find TP, FP, TN, FN, Accuracy, Error rate, Precision, Recall on the given dataset.

# 📦 Step 1: Import Required Libraries
import pandas as pd                      # For handling dataset
import numpy as np                       # For numerical computations
import matplotlib.pyplot as plt          # For plotting graphs
import seaborn as sns                    # For advanced plotting
from sklearn.linear_model import LogisticRegression  # For logistic regression model
from sklearn.model_selection import train_test_split # To split data into training and testing
from sklearn.preprocessing import StandardScaler      # To scale the features
from sklearn.metrics import confusion_matrix, classification_report  # To evaluate model

# 📂 Step 2: Load Dataset
df = pd.read_csv("Social_Network_Ads.csv")  # Load CSV file into DataFrame

# 👀 View top records and structure
print(df.head())      # Display first 5 rows
print(df.info())      # Show column info and data types

# 🧹 Data Preprocessing
# A. Check for Missing Values
print("\nMissing Values:\n", df.isnull().sum())  # Check for null values in each column

# Drop rows with missing values if any
df_before = df.copy()         # Save original copy
df = df.dropna()              # Remove rows with nulls

# Compare size before and after dropping missing values
print(f"Before dropping rows: {df_before.shape}")
print(f"After dropping rows: {df.shape}")

# 📊 Descriptive Statistics
print(df.describe())          # Show summary statistics like mean, std dev, min, max

# 🔍 Check Data Types
print(df.dtypes)              # Show data type of each column

# 🧪 Step 2: Outlier Detection using IQR (Interquartile Range)
numeric_cols = ['Age', 'EstimatedSalary']  # Columns to check outliers
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)                    # 25th percentile
    Q3 = df[col].quantile(0.75)                    # 75th percentile
    IQR = Q3 - Q1                                  # IQR range
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]  # Detect outliers
    print(f"\nOutliers in {col}:\n", outliers[[col]])  # Print outliers

# ℹ️ If the result shows "Empty DataFrame", it means no outliers are found.

# ⚖️ Step 3: Feature Scaling using StandardScaler
X = df[['Age', 'EstimatedSalary']]  # Select features
y = df['Purchased']                 # Select target column

# Normalize the feature values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Scale data

# Convert scaled data to DataFrame for inspection
X_scaled_df = pd.DataFrame(X_scaled, columns=['Age', 'EstimatedSalary'])
print(X_scaled_df.head())          # View scaled features

# 🧠 Step 4: Logistic Regression Model Building

# A. Split the data into training and testing sets (70-30 ratio)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# B. Create and train Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)        # Train the model

# C. Make predictions
y_pred = model.predict(X_test)     # Predict labels for test set

# 📈 Step 5: Evaluate Model using Confusion Matrix

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)

# Extract TN, FP, FN, TP from confusion matrix
TN, FP, FN, TP = cm.ravel()

# Calculate Accuracy, Error Rate, Precision, Recall
accuracy = (TP + TN) / (TP + TN + FP + FN)       # Correct predictions / total
error_rate = 1 - accuracy                        # Wrong predictions / total
precision = TP / (TP + FP)                       # True Positive / Predicted Positive
recall = TP / (TP + FN)                          # True Positive / Actual Positive

# Display metrics
print(f"\nTP: {TP}, FP: {FP}, TN: {TN}, FN: {FN}")
print(f"Accuracy: {accuracy:.2f}")
print(f"Error Rate: {error_rate:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")

# 🎨 Step 6: Visualize Confusion Matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')  # Heatmap with annotations
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
